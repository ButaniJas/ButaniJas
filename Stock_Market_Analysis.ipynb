###IMPORT LIBRARIES

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

###Fetch Stock Data

```python
import yfinance as yf
import pandas as pd

def fetch_nifty_data():
    nifty = yf.Ticker("^NSEI")
    data = nifty.history(period="7d", interval="1h")
    data.reset_index(inplace=True)
    return data

nifty_data = fetch_nifty_data()
nifty_data.head()
```

###Line Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(nifty_data['Datetime'], nifty_data['Close'], label='Close Price', color='red')
plt.xlabel("Date")
plt.ylabel("Nifty 50 Index Value")
plt.title("Nifty 50 Closing Price Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

###Candle-Stick Chart

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=nifty_data['Datetime'],
                open=nifty_data['Open'], high=nifty_data['High'],
                low=nifty_data['Low'], close=nifty_data['Close'])])
fig.update_layout(title="Nifty 50 Index (Candlestick Chart)",
                  xaxis_title="Date", yaxis_title="Index Value")
fig.show()
```

###Moving Averages(SMA, EMA)

```python

nifty_data['SMA_50'] = nifty_data['Close'].rolling(window=50).mean()
nifty_data['EMA_20'] = nifty_data['Close'].ewm(span=20, adjust=False).mean()

plt.figure(figsize=(10, 5))
plt.plot(nifty_data['Datetime'], nifty_data['Close'], label='Close Price', color='blue')
plt.plot(nifty_data['Datetime'], nifty_data['EMA_20'], label='20-day EMA', color='green')
plt.xlabel("Date")
plt.ylabel("Nifty 50 Index Value")
plt.title("Nifty 50 with Moving Averages")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
